# Sentiment / Subjectivity Analysis

```python
from textblob import TextBlob

def sentiment_calc(text):
    try:
        return TextBlob(text).sentiment
    except:
        return None
    
df['sentiment']=df['Contents'].apply(sentiment_calc)
df
```

```python
new_col_list = ['sentiment score ','subjectivity']
for n,col in enumerate(new_col_list):
    df[col] = df['sentiment'].apply(lambda location: location[n])
df=df.drop('sentiment', axis=1)

df.describe()
```

# Abusive Words

```python
words=pd.read_excel('Abusive Words List.xlsx')
words.head()
```

```python
words['Key Word']=words['Key Word'].apply(lambda x: x if type(x)!=str else x.lower()) # lower case
words['Key Word']=words['Key Word'].apply(lambda x: ''.join([i for i in x
                                                                  if i not in string.punctuation]))
```

```python
import re

def is_present(text, key_words):
    non_alphanum='[^A-Za-z0-9]'
    count=0
    for phrase in key_words:
        r= re.compile(non_alphanum + phrase + non_alphanum)
        match=re.search(r, text)
        if match:
            count +=1
    return count
```

```python
df['AbuseCount'] = df['Contents'].map(lambda x: is_present(x, words['Key Word']))
```

```python
abusiveDF=df[(df['AbuseCount']!=0)]
```
