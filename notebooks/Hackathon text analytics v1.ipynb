```python
import pandas as pd
import nltk
import string
from nltk.corpus import stopwords
```

```python
june01DF=pd.read_excel('06-01.xls')
june01DF['Contents']=june01DF['Contents'].apply(lambda x: x if type(x)!=str else x.lower()) # lower case
june01DF['Contents']=june01DF['Contents'].apply(lambda x: ''.join([i for i in x
                                                                  if i not in string.punctuation]))
#june01DF['Contents']=june01DF['Contents'].apply(nltk.word_tokenize)

#stop = stopwords.words('english')


#june01DF['Contents']=june01DF['Contents'].apply(lambda x: ''.join([i for i in x 
#                                                       if i not in stop]))
june01DF
```

```python
from textblob import TextBlob
```

```python
def sentiment_calc(text):
    try:
        return TextBlob(text).sentiment
    except:
        return None
    
june01DF['sentiment']=june01DF['Contents'].apply(sentiment_calc)
june01DF
```

```python
june01DF['sentiment1', 'subjectivity']=june01DF['sentiment'].apply(eval).apply(pd.Series)
```

```python
new_col_list = ['sentiment score ','subjectivity']
for n,col in enumerate(new_col_list):
    june01DF[col] = june01DF['sentiment'].apply(lambda location: location[n])
june01DF=june01DF.drop('sentiment', axis=1)
```
